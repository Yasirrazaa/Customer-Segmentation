```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("OnlineRetail.csv",encoding='latin-1')
```

```python

df.dropna(inplace=True)
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df["CustomerID"]=df["CustomerID"].astype('int64')

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'],format='%d-%m-%Y %H:%M')
df["CustomerID"]=df["CustomerID"].astype(str)
```

```python
df.info()
```

```python
df.info()
```

```python
df['Amount']=df['Quantity']*df['UnitPrice']
df
```

### Let's Calculate
#####    Monetary
#####   Frequency
#####    Recency

```python
monetary=df.groupby('CustomerID')['Amount'].sum()
monetary=monetary.reset_index()
monetary
```

```python
frequency=df.groupby('CustomerID')['InvoiceNo'].count()
frequency=frequency.reset_index()
frequency.columns=['CustomerID','Frequency']
frequency
```

```python
mon_freq=pd.merge(monetary,frequency,on='CustomerID',how='inner')
mon_freq
```

```python
max_date=max(df['InvoiceDate'])
df['date_diff']=(max_date-df['InvoiceDate']).dt.days
df
```

```python
recent=df.groupby('CustomerID')['date_diff'].min()
recent=recent.reset_index()
recent.columns=['CustomerID','Recency']
recent
```

```python
final=pd.merge(mon_freq,recent,on='CustomerID',how='inner')
final
```

## Detecting Outliers and Removing them

```python
# Outlier Analysis
columns=['Amount','Frequency','Recency']
plt.figure(figsize=(10,8))
sns.boxplot(data=final[columns],orient='v',palette='Set2',whis=1.5,saturation=1,width=0.7)
plt.title("Outlier Analysis",fontsize=20)
plt.ylabel("Range",fontsize=15)
plt.xlabel("Attributes",fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Removing Outliers
# Amount
Q1=final.Amount.quantile(0.05)
Q3=final.Amount.quantile(0.95)
IQR=Q3-Q1
final=final[(final.Amount>=Q1-1.5*IQR)&(final.Amount<=Q3+1.5*IQR)]
# Frequency
Q1=final.Frequency.quantile(0.05)
Q3=final.Frequency.quantile(0.95)
IQR=Q3-Q1
final=final[(final.Frequency>=Q1-1.5*IQR)&(final.Frequency<=Q3+1.5*IQR)]
# Recency
Q1=final.Recency.quantile(0.05)
Q3=final.Recency.quantile(0.95)
IQR=Q3-Q1
final=final[(final.Recency>=Q1-1.5*IQR)&(final.Recency<=Q3+1.5*IQR)]
```

```python
# Outlier Analysis
columns=['Amount','Frequency','Recency']
plt.figure(figsize=(10,8))
sns.boxplot(data=final[columns],orient='v',palette='Set2',whis=1.5,saturation=1,width=0.7)
plt.title("Outlier Analysis",fontsize=20)
plt.ylabel("Range",fontsize=15)
plt.xlabel("Attributes",fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

## Feature Scaling

```python
to_scale=['Amount','Frequency','Recency']
scale=StandardScaler()
scaled=scale.fit_transform(final[to_scale])
scaled=pd.DataFrame(scaled)
scaled.columns=['Amount','Frequency','Recency']
scaled
```

## Model Building

```python
# K-Means Clustering
# Elbow Curve
sse={}
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=1)
    kmeans.fit(scaled)
    sse[k]=kmeans.inertia_
plt.figure(figsize=(10,8))
plt.title('The Elbow Curve')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()),y=list(sse.values()))
plt.show()
```

```python
# n_clusters=3
kmeans=KMeans(n_clusters=3,max_iter=500)
kmeans.fit(scaled)
final['clusters']=kmeans.labels_
```
